```python
# Import pandas library
import pandas as pd

# Read the Excel file into a dataframe
df = pd.read_excel ('Assignment_Timecard.xlsx')

# Convert the date column to datetime format
df ['Pay Cycle Start Date'] = pd.to_datetime (df ['Pay Cycle Start Date'])

# Group the dataframe by name and position
grouped = df.groupby (['Employee Name', 'Position ID'])

# Define a function to check if a series has 7 consecutive days
def has_consecutive_days (series):
    # Get the difference between each date and the previous date
    diff = series.diff ()
    # Count how many times 1 day difference occurs
    count = (diff == pd.Timedelta ('1 day')).cumsum ()
    # Check if there is any 7 day streak
    return count.groupby (count).transform ('size').ge (7).any ()

# Apply the function to each group and filter the ones that return True
filtered = grouped.filter (lambda x: has_consecutive_days (x ['Pay Cycle Start Date']))

# Drop duplicates by name and position
filtered = filtered.drop_duplicates (['Employee Name', 'Position ID'])

# Print the name and position of employees who have worked for 7 consecutive days
print ('Employees who have worked for 7 consecutive days:')
for index, row in filtered.iterrows ():
    print (row ['Employee Name'], '-', row ['Position ID'])
```
